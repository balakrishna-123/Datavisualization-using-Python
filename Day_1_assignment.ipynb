```python
import numpy as np
import pandas as pd
import seaborn as sns

import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# plot a line plot between a and b:
a=np.arange(40,50)
b=np.arange(50,60)
plt.title('line plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.plot(a,b)
```

```python
#Plot a line plot showing the sales trend in company 1 and 2:

days = [1,2,3,4,5,6,7] #days of d week
sales_1 = [160,150,140,145,175,165,180] #sales of company1
sales_2 = [70,90,160,150,140,145,175]  #sales of company2
plt.plot(days,sales_1 ,label='sales_1')
plt.plot(days,sales_2 ,label='sales_2')
plt.legend()
```

```python
# Create a 3 by 3 subplots multiple plots 
x = [1,2,3,4]
y1 = [4,3,2,1]
y2 = [10,20,30,40]
y3 = [40,30,20,10]
y4 = [1,2,1,2]
y5 = [40,70,90,70]
plt.subplot(3,3,1)
plt.plot(x,y1,'go')
plt.subplot(3,3,2)
plt.plot(x,y2,'bo')
plt.subplot(3,3,3)
plt.plot(x,y3,'ro')
plt.subplot(3,3,4)
plt.plot(x,y4,'g*-')
plt.subplot(3,3,5)
plt.plot(x,y5,'r--o')
```
